```python
while True:
    trigPin = Pmod_IO(PMODA,3,"out")

    analog1 = Arduino_IO(3,0,"in")

    trigPin.write(0)
    time.sleep(0.5)

    trigPin.write(1)
    time.sleep(0.00001)
    trigPin.write(0)
    y = 0
    for i in range (0, 10):
        x = analog1.read()
        print(x)
        y = y + x
    print("this is y", y)
    print("another round")
    time.sleep(2)
```
